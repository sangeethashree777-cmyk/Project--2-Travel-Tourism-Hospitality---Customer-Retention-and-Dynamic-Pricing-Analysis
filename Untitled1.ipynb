```python
df=pd.read_csv(r"C:\Users\Harish\Downloads\Project -2 Travel-Tourism-Hospitality---Customer-Retention-and-Dynamic-Pricing-Analysis\Huge_Uncleaned_Travel_Tourism_Hospitality.csv")
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Destination']=df['Destination'].ffill()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Number_of_Guests']=df['Number_of_Guests'].bfill()
```

```python
df.isnull().sum()
```

```python
df['Customer_Rating']=df['Customer_Rating'].interpolate()
```

```python
df.isnull().sum()
```

```python
df['Discount_Percentage'] = df['Discount_Percentage'].fillna(df['Discount_Percentage'].median())
```

```python
df.isnull().sum()
```

```python
import numpy as py
```

```python
import seaborn as sns
```

```python
import matplotlib as plt
```

```python
df['Lead_Time_Days'] = df['Lead_Time_Days'].fillna(df['Lead_Time_Days'].median())

df['Review_Score'] = df['Review_Score'].fillna(df['Review_Score'].median())
```

```python
df.isnull().sum()
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna(df['Customer_Name'].mode()[0])

df['Customer_Type'] = df['Customer_Type'].fillna(df['Customer_Type'].mode()[0])

df['Age_Group'] = df['Age_Group'].fillna(df['Age_Group'].mode()[0])

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

df['Country'] = df['Country'].fillna(df['Country'].mode()[0])

df['Room_Type'] = df['Room_Type'].fillna(df['Room_Type'].mode()[0])

df['Agent_Name'] = df['Agent_Name'].fillna(df['Agent_Name'].mode()[0])

df['Loyalty_Member'] = df['Loyalty_Member'].fillna(df['Loyalty_Member'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.to_csv("Cleaned_Travel_Tourism_Hospitality.csv",index=False)
```
